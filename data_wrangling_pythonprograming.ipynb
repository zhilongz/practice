Tutorial on data analysis in Python -- pythonprogramming.net 

```python
import pandas as pd 
from pandas import DataFrame
import datetime
```

Get data from SP500 and save it to CSV 

```python
import pandas.io.data
sp500 = pd.io.data.get_data_yahoo('%5EGSPC',start=datetime.datetime(2000,10,1),end=datetime.datetime(2012,1,1))
```

```python
print sp500.head() #print the first 5 for visualization
```

```python
sp500.to_csv('sp500_ohlc.csv') # save df to csv
```

```python
df = pd.read_csv('sp500_ohlc.csv',index_col='Date',parse_dates = True) # converts the date into date obj
df.head()
```

DF column manipulation 

```python
df2 = df['Open'] # select a column knowning its column name 
print df2.head()
```

```python
df3 = df[['Close','High']] # multiple columns 
print df3.head()
```

```python
df3.rename(columns={'Close':'Close!'},inplace=True) # rename a column, inplace means modify the column inplace w/o new column
print df3.head()
```

```python
df4 = df3[(df3['Close!']>1400)] # apply filter to a column of data
print df4.head()
```

```python
# delete a column 2 methods 
df5 = df4.drop('Close!',1) # 0 for rows and 1 for columns
print df5.head()
```

```python
del df4['Close!']
print df4.head()
```

```python
df6 = df
df6.drop(df.columns[[1,2,4]],axis=1) #drop based on pd column index, first column is 0
```

panda basic maths operation

```python
df['H-L'] = df.High - df.Low # add a column with some basic maths. Here, we accessed columns like objects 
print df.head()
```

```python
df['t100MA'] = pd.rolling_mean(df['Close'],100) # calculate the moving average in 100 points 
print df.t100MA[200:210]
# print df[200:210]
```

```python
df['Difference'] = df.Close.diff() # computes the difference from previous value, very useful feature
print df.head()
```

visualization with Panda

```python
import matplotlib.pyplot as plt #module for plotting
```

```python
df[['Open','High','Low','Close','t100MA']].plot() # what's great about this is that I don't have to mess at all with the axis and dates 
plt.show() # pop up in a different window
# panda offers a custom version of Matplotlib 
```

Statistics with Pandas 

```python
df['STD'] = pd.rolling_std(df.Close,25,min_periods=1)
```

```python
ax1 = plt.subplot(2,1,1)
df.Close.plot()
ax2 = plt.subplot(2,1,2, sharex = ax1)
df.STD.plot()
plt.show()
```

Advanced: overcome tools not predefined in panda 

```python
import random
def my_function(data):
    x = random.randrange(0,5)
    return data*x 
```

```python
df['Multiple']=map(my_function, df.Close) # map functions avoids iteration, which slows down panda
print df.head()
```
